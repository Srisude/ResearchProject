```python
from collections import defaultdict

def categorize_labels(label_list, category_keywords):
    categorized = defaultdict(list)

    for label in label_list:
        matched = False
        for category, keywords in category_keywords.items():
            for keyword in keywords:
                if keyword in label.lower():
                    categorized[category].append(label)
                    matched = True
                    break
            if matched:
                break
        if not matched:
            categorized["UNCATEGORIZED"].append(label)

    return categorized
```

```python
import json
import os
import pandas as pd

# JSON dosyasını yükle
json_path = "/content/kilogram/dataset/full.json"
with open(json_path, "r") as f:
    data = json.load(f)

# Görsel-caption eşleşmeleri
image_caption_pairs = []

for img_name, img_data in data.items():
    annotations = img_data.get("annotations", [])
    if annotations:
        caption = annotations[0]['whole'].get('wholeAnnotation', None)
        if caption:
            filename = f"{img_name}.svg"
            image_caption_pairs.append((filename, caption))

# DataFrame olarak kaydet
df = pd.DataFrame(image_caption_pairs, columns=["filename", "caption"])
df.to_csv("tangram_caption_dataset.csv", index=False)

print(f"✅ {len(df)} tane eşleşme CSV dosyasına yazıldı.")
df.head(10)
```

```python
# Görsel adı -> caption sayısını say
from collections import defaultdict

multi_caption_images = defaultdict(list)

for img_name, img_data in data.items():
    annotations = img_data.get("annotations", [])
    for ann in annotations:
        caption = ann['whole'].get('wholeAnnotation', None)
        if caption:
            multi_caption_images[img_name].append(caption)

# Birden fazla caption'ı olanları filtrele
multi_caption_images = {k: v for k, v in multi_caption_images.items() if len(v) > 1}

# Örnek olarak ilk 5 tanesini göster
for i, (img, captions) in enumerate(multi_caption_images.items()):
    print(f"📄 {img}.svg")
    for idx, c in enumerate(captions):
        print(f"   {idx+1}. {c}")
    print()
    if i >= 4:
        break
```

```python
import re
from collections import defaultdict

# Etiketlerin listesi (label'lar burada unique_labels gibi olmalı)
# örnek olarak: unique_labels = list(set(labels))

# Temel keyword listesi
keywords = [
    "cat", "dog", "person", "car", "house", "robot", "bird",
    "table", "giraffe", "rabbit", "snake"
]

# Belirli keyword'lar için blacklist (yanlış eşleşmeleri engelle)
blacklist = {
    "car": ["carrot", "carousel", "carrying", "baby carseat"],
    "cat": ["caterpillar", "meercat"],
    "dog": ["hotdog"],
    "bird": ["birdhouse"]
}

# Fonksiyon: Etikette keyword tam kelime olarak geçiyor mu?
def keyword_in_label(label, keyword):
    return bool(re.search(r'\b' + re.escape(keyword) + r'\b', label.lower()))

# Sonuçları saklamak için dict
keyword_hits = defaultdict(list)

# Etiketleri keyword'lara göre sınıflandır
for label in unique_labels:
    for keyword in keywords:
        if keyword_in_label(label, keyword):
            # Blacklist kontrolü
            if any(blk in label.lower() for blk in blacklist.get(keyword, [])):
                continue
            keyword_hits[keyword].append(label)

# Sonuçları yazdır
for keyword, matches in keyword_hits.items():
    print(f"\n📁 {keyword.upper()} ({len(matches)} eşleşme):")
    for i, match in enumerate(matches[:10], 1):  # ilk 10 örnek göster
        print(f"{i}. {match}")
    if len(matches) > 10:
        print(f"... (+{len(matches) - 10} tane daha)")
```

```python
import json
from IPython.display import SVG, display

# JSON dosyasını yükle
with open("/content/kilogram/dataset/full.json") as f:
    data = json.load(f)

# Aranan açıklama
target_label = "snake behind a rock"

# Eşleşen dosyayı bul
matching_files = []
for filename, content in data.items():
    for ann in content["annotations"]:
        if ann["whole"]["wholeAnnotation"].lower() == target_label.lower():
            matching_files.append(filename)

print("Eşleşen dosyalar:", matching_files)

# İlk eşleşmeyi göster
if matching_files:
    file_path = f"/content/kilogram/dataset/tangrams-svg/{matching_files[0]}.svg"
    display(SVG(filename=file_path))
else:
    print("Görsel bulunamadı.")
```
